```python
import matplotlib.pyplot as plt
```

```python
#cara menampilkan grafik BAR
fig = plt.figure()
ax = fig.add_axes([0,0,1,1,])
jurusan = ['RPL','PH','OTKP']
siswa = [130, 90, 120]
c = ['blue','green','red']
plt.title("PPDB SMK AL MADANI GARUT")
plt.bar(jurusan, siswa, color = c)
plt.show()
```

```python
#pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("Data siswa PPDB SMK AL MADANI GARUT")
jurusan = ['RPL','PH','OTKP']
siswa = [130, 90, 120]
ax.pie(siswa,labels = jurusan,autopct='%1.2f%%')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2,])
nama = ['Tia','Hana','Widia','Romi','Samsul','Dimas']
belanjaan = [246600, 193200.0, 119300, 221700, 323800, 379300, ]
c = ['blue','green','red','orange','black','yellow']
plt.title("PPDB SMK AL MADANI GARUT")
plt.bar(nama, belanjaan, color = c)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
nama = ['Tia','Hana','Widia','Romi','Samsul','Dimas']
belanjaan = [246600, 193200.0, 119300, 221700, 323800, 379300, ]
siswa = [130, 90, 120]
ax.pie(belanjaan,labels = nama,autopct='%1.2f%%')
plt.show()
```
